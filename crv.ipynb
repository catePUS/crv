```python
import math
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
d = {3:4, 0:1, 1:8, 9:3, 2:9}
for i,k in d.items():
    sigma=math.sqrt(k)
    x = np.linspace(i - 3*sigma, i + 3*sigma, 100)
    plt.plot(x, scipy.stats.norm.pdf(x, i, sigma))
plt.show()    
    
    
```

```python
d = {3:4, 0:1, 1:8, 9:3, 2:9}

def drawChart(i, k, plt): 
   sigma= math.sqrt(k)
   x = np.linspace(i - 3*sigma, i + 3*sigma, 100)
   plt.plot(x, scipy.stats.norm.pdf(x, i, sigma))
for i, k in d.items(): 
  drawChart(i, k, plt)
plt.show()
```
